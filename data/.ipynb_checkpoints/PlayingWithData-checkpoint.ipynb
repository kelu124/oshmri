```python
import numpy as np
import matplotlib.pyplot as plt
  
```

```python
melon = np.load("raw kspace - Melon.npy")
```

```python
plt.imshow(np.abs(melon[64]), aspect=1)
```

```python
plt.figure(figsize=(15,5))
plt.subplot(1, 3, 1)
plt.imshow(np.abs(np.fft.fftshift(np.fft.ifft2(melon[63]))), aspect=1)
plt.subplot(1, 3, 3)
plt.imshow(np.abs(np.fft.fftshift(np.fft.ifft2(melon[:,:,63]))), aspect=1)
plt.subplot(1, 3, 2) 
plt.imshow(np.abs(np.fft.fftshift(np.fft.ifft2(melon[:,63]))), aspect=1)
plt.savefig("melon.png")
plt.show()
```
